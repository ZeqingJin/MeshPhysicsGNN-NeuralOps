```python
import numpy as np
import pandas as pd
import os
from sklearn.metrics.pairwise import euclidean_distances
```

```python
data_file='/content/drive/MyDrive/hyundai_collaboration/Code_summary/neural_operator/example_csv_data'


def new_connectivity(data_file):
    data_list=[]
    file_name_list=[]
    file_list=os.listdir(data_file)
    k=1
    for item in file_list:
        design_name=item.rpartition('_')[0]

        if design_name != ".DS" and design_name not in file_name_list:
            print(k)
            design_node=data_file+'/'+design_name+'_nodes.csv'
            df = pd.read_csv(design_node)

            point_cloud=df.loc[:,['x','y','z']].to_numpy()
            connectivity=euclidean_distances(point_cloud,point_cloud)

            df_connectivity = pd.DataFrame(connectivity)
            output_file = pd.concat([df, df_connectivity], axis=1)
            output_file.to_csv(data_file+'/'+design_name+'_Nodes.csv',index=False)

            file_name_list.append(design_name)
            k=k+1

new_connectivity(data_file)
```
